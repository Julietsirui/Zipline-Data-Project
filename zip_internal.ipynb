```python
import csv

import pandas as pd
from pandas import DataFrame

import matplotlib.pyplot as plt

import numpy as np

import seaborn as sb
```

```python
with open('flight_17678.csv', 'r') as csvfile:
    flightCSV = csv.reader(csvfile, delimiter = ',')
    
flight = pd.read_csv("flight_17678.csv", error_bad_lines=False, header = None, delimiter = ',')
    
flight.columns = ["seconds since launch", "position(n)","position(e)", "position(d)", "v(n)","v(e)","v(d)", "acc(forward)","acc(rightward)", "acc(down)", "orientation(euler roll)","orientation(euler pitch)","orientation(euler yaw)", "angular v(for)","angular v(right)","angular v(down)", "position sigma(north)","position sigma(east)", "position sigma(down)"]
flight = flight.convert_objects(convert_numeric = True)
```

```python
flight_sim = flight[['seconds since launch', 'position(d)', 'angular v(down)', 'orientation(euler yaw)', 'position sigma(down)']]
```

```python
plt.plot(flight['v(d)'], flight['position sigma(down)'], 'o', c = 'dodgerblue')
plt.xlabel('velocity')
plt.ylabel('position sigma')
plt.plot(flight['v(n)'], flight['position sigma(north)'], 'o', c = 'red')
plt.plot(flight['v(e)'], flight['position sigma(east)'], 'o', c = 'yellow')
plt.legend()
```
